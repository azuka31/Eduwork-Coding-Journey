```python
df = pd.read_csv('healtcare_stroke_data.csv')
df.head()
```

```python
def categorize_age(age):
    if age <= 35:
        return 'Young'
    elif 36 <= age <= 60:
        return 'Middle-aged'
    else:
        return 'Senior'
```

```python
df['age_category'] = df['age'].apply(categorize_age)
```

```python
df.head()
```

```python
residence_counts = df.groupby('Residence_type').size().to_dict()
```

```python
print(residence_counts)
```

```python
def get_stroke_risk(row):
    if row['hypertension'] == 1 and row['avg_glucose_level'] > 150:
        return 'High'
    else:
        return 'Low'
```

```python
df['stroke_risk'] = df.apply(get_stroke_risk, axis=1)
```

```python
df.head()
```

```python
# Using boolean indexing to filter rows
filtered_df = df[(df['age'] > 60) & (df['bmi'] < 25) & (df['smoking_status'] == 'never smoked')]

# Printing the IDs of individuals who meet the criteria
for id_value in filtered_df['id']:
    print(f"ID: {id_value} matches the criteria!")
```

```python
def categorize_glucose_lv(avg_glucose_level):
    if avg_glucose_level <= 100:
        return 'Normal'
    elif 101 <= avg_glucose_level <= 125:
        return 'Prediabetes'
    else:
        return 'Diabetes'
```

```python
df['diabetes_category'] = df['avg_glucose_level'].apply(categorize_glucose_lv)
df.head()
```
