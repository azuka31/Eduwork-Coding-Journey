# Import Preparations

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_summarysales = pd.read_csv('../data/summary_sales.csv', sep='|')
```

```python
df_summarysales
```

# Line Chart

```python
# X bawah, Y kiri
```

```python
# Di sini kita panggil masing-masing garis ProductA, ProductB, dan ProductC (dg df masing-masing yA, yB, yC)
xA = df_summarysales['Month'].values
yA = df_summarysales['ProductA'].values
```

```python
xB = df_summarysales['Month'].values
yB = df_summarysales['ProductB'].values
```

```python
xC = df_summarysales['Month'].values
yC = df_summarysales['ProductC'].values
```

```python
# DISCLAIMER: SEHARUSNYA x axis nya bisa sekali saja tidak usah dipisah xA xB xC
```

```python
# figure figsize di sini (panjang, tinggi)
plt.figure(figsize=(8, 6))

# plt.plot utk memanggil / menggambar chart, utk line chart bisa multiple line yang kita plot, tinggal di tulis ulang line baru
# plot di sini seperti css, bisa beri semacam "style", marker itu bentuk apa seperti titik dll, linestyle contoh bisa - lurus atau -- garis putus"
# yang menentukan mau menampilkan apa itu 2 parameter awal (contoh xA, yA & xB, yB atau xC, yC, dst)
# label = 'di sini utk beri nama garis yg bisa ditunjukkan dg plt.legend'
plt.plot(xA, yA, marker='o', linestyle='-', label='ProductA')
plt.plot(xB, yB, marker='o', linestyle='-', label='ProductB')
plt.plot(xC, yC, marker='o', linestyle='-', label='ProductC')

# Beri label x (bawah) dan y (kiri)
plt.xlabel('Month')
plt.ylabel('Product Amount')

# xticks di sini utk rotasi label di axis X (kalau mau Y bisa yticks)
plt.xticks(rotation=45)

# beri title di atas
plt.title('Sales Monthly Product by Products')

# plt.legend utk tunjukkan label masing-masing garis || plt.show utk menunjukkan hasil akhir
plt.legend() 
plt.show()
```

# Bar Chart

```python
# Di sini kita bikin df X_Month dg value pada kolom Month
X_Month = df_summarysales['Month'].values

# Lalu kita buat axis yA, yB, yC (Bisa lanjut sebenarnya) untuk ProductA, B, dan C
yA = df_summarysales['ProductA'].values
yB = df_summarysales['ProductB'].values
yC = df_summarysales['ProductC'].values

#Ini seharusnya utk menentukan lebar X axis agar bisa menampung keduabelas bulan (Values pada kolom Month), kita beri nama df nya sbg X_axis
X_axis = np.arange(len(X_Month))
```

```python

#REFERENCE: https://www.scaler.com/topics/matplotlib/plotting-multiple-bar-chart/

# figure figsize di sini (panjang, tinggi)
plt.figure(figsize=(8, 6))

# Di sini kita buat 3x bar utk masing" ProdukA,ProdukB,ProdukC 
# plt.bar utk membuat barchart (X_axis di sini memanggil df di atas yg kolom Month)
# -0.3 dan +0.3 ini posisi BarChart nya, karena ada 3 bar maka posisi yB tidak usah diubah stay di 0 karena di tengah
# width di sini lebar bar nya, kita kecilkan agar tidak bentrok karena ada 12 bulan
# label utk di legend nanti, color utk warna bar nya
plt.bar(X_axis -0.3, yA, width=0.2, 
        label = 'Product A', color='r')
plt.bar(X_axis, yB, width=0.2, 
        label = 'Product B', color='y')
plt.bar(X_axis +0.3, yC, width=0.2, 
        label = 'Product C', color='g')

# Xticks
plt.xticks(X_axis, X_Month, rotation=45)

# beri title di atas
plt.title('Sales Monthly Product by Products')

plt.legend() 
plt.show()
```

# Heatmap

```python
df_summarysales = pd.read_csv('../data/summary_sales.csv', sep='|')

# Set kolom Month sbg index
df_summarysales.set_index('Month', inplace=True)

# Pembuatan heatmap dengan Seaborn library
plt.figure(figsize=(14, 6))
#annot=true utk display value cell yg tidak sama (kalau sama atau berulang tdk dishow dua kali)
#fmt="g" utk format angka nya, contoh kalau fmt=".2f" itu dua angka di belakang koma
#cmap='viridis' itu template warnanya, katanya viridis cukuup populer
sns.heatmap(df_summarysales, annot=True, fmt="g", cmap='viridis')
#plt.title('Sales Monthly Product by Products')
plt.show()
```
