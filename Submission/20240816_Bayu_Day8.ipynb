```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df_sales_data = pd.read_csv('C:\\Users\\Bayu\\anaconda3\\Lib\\site-packages\\pandas\\io\\parsers\\summary_sales.csv' , sep='|')
```

```python
df_sales_data = df_sales_data.loc[:, ~df_sales_data.columns.str.contains('^Unnamed')]
```

#Choose product =

```python
print(df_sales_data[['Month', 'ProductA', 'ProductG']])
```

```python
#Bar Chart
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
#Data Preparing
x = df_sales_data['Month']
yA= df_sales_data['ProductA']
yG= df_sales_data['ProductG']
x = np.arange(len(df_sales_data['Month']))
width = 0.35
fig, ax = plt.subplots(figsize=(12, 8))



#Plot bar for A and G
bars1 = ax.bar(x - width/2, df_sales_data['ProductA'], width, label='ProductA', color='blue', alpha=0.7)
bars2 = ax.bar(x + width/2, df_sales_data['ProductG'], width, label='ProductG', color='green', alpha=0.7)

#addition text
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.set_title('Sales of ProductA and ProductG by Month')
ax.set_xticks(x)
ax.set_xticklabels(df_sales_data['Month'], rotation=45)
ax.legend()

#show plot
plt.tight_layout
plt.show()
```

```python
#Line chart
plt.figure(figsize=(12, 8))
plt.plot(df_sales_data['Month'], df_sales_data['ProductA'], marker='o', label='ProductA', color='blue', linestyle='-', linewidth=2)
plt.plot(df_sales_data['Month'], df_sales_data['ProductG'], marker='o', label='ProductG', color='green', linestyle='-', linewidth=2)

#addition text
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales of ProductA and ProductG')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.legend()

#Show plot
plt.tight_layout()
plt.show()
```
